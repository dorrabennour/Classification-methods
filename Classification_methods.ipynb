```
def LOOCV_RDF(X,y):
    # create loocv procedure
    cv = LeaveOneOut()
    # enumerate splits
    y_true, y_pred = list(), list()
    for train_ix, test_ix in cv.split(X):
        # split data
        X_train, X_test = X[train_ix, :], X[test_ix, :]
        y_train, y_test = y[train_ix], y[test_ix]
        # fit model
        model_LOO = RandomForestClassifier(random_state=0)
        model_LOO.fit(X_train, y_train)
        # evaluate model
        yhat = model_LOO.predict(X_test)
        # store
        y_true.append(y_test[0])
        y_pred.append(yhat[0])
    # calculate accuracy
    acc_LOO = accuracy_score(y_true, y_pred)
    Y_score = model_LOO.predict_proba(X_test)[:,1]
    #print('Accuracy: %.3f' % acc_LOO)
    return acc_LOO,y_true,y_pred,Y_score,model_LOO
```

```
# Leave one out cross validation for k-nearest neighbors 
def LOOCV_KNN(X,y,k):
    # create loocv procedure
    cv = LeaveOneOut()
    # enumerate splits
    y_true, y_pred = list(), list()
    for train_ix, test_ix in cv.split(X):
        # split data
        X_train, X_test = X[train_ix, :], X[test_ix, :]
        y_train, y_test = y[train_ix], y[test_ix]
        # fit model
        model_LOO = KNeighborsClassifier(n_neighbors = k)
        model_LOO.fit(X_train, y_train)
        # evaluate model
        yhat = model_LOO.predict(X_test)
        # store
        y_true.append(y_test[0])
        y_pred.append(yhat[0])
    # calculate accuracy
    acc_LOO = accuracy_score(y_true, y_pred)
    Y_score = model_LOO.predict_proba(X_test)[:,1]
    #print('Accuracy: %.3f' % acc_LOO)
    return acc_LOO,y_true,y_pred,Y_score,model_LOO
```
